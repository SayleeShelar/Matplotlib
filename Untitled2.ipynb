# Bar Plot

```python
import matplotlib.pyplot as plt  # Import the pyplot module from matplotlib for creating visualizations.

x = ["python", "C", "C++", "Java"]  # Define the categories for the x-axis, which are different programming languages.
y = [85, 70, 60, 82]  # Define the data for the y-axis, which represents marks or scores for each programming language.

plt.xlabel("Programming Language", fontsize=10)  # Set the label for the x-axis with a font size of 10.
plt.ylabel("Marks")  # Set the label for the y-axis as "Marks".
plt.title("Student Performance")  # Set the title of the bar graph to "Student Performance".

c = ["b", "g", "c", "m"]  # Define a list of colors for the bars: blue, green, cyan, and magenta.
plt.bar(x, y, width=0.4, color=c, align="center", edgecolor="r", linewidth=2, linestyle=":", alpha=0.9, label="Student Performance")
# Create a bar graph with the following parameters:
# - x: the categories (programming languages) for the x-axis.
# - y: the data (marks) for the y-axis.
# - width: the width of each bar.
# - color: the colors of the bars.
# - align: alignment of the bars to the center of the ticks.
# - edgecolor: color of the edges of the bars (red in this case).
# - linewidth: width of the edges of the bars.
# - linestyle: style of the edges (dotted in this case).
# - alpha: transparency level of the bars (0.9 means slightly transparent).
# - label: label for the legend.

plt.legend()  # Add a legend to the graph, using the label specified in the plt.bar() call.
plt.show()  # Display the bar graph.
```

```python
import matplotlib.pyplot as plt  # Import the plotting library for visualizations.
import numpy as np  # Import NumPy for numerical operations and array handling.

# Define the categories (programming languages) for the x-axis.
x = ["python", "C", "C++", "Java"]

# Define the data sets (marks or scores) for the y-axis:
y = [85, 70, 60, 82]  # First dataset, e.g., marks in a popularity test.
z = [20, 30, 40, 50]  # Second dataset, e.g., marks in another test or a different metric.

width = 0.2  # Set the width of each bar in the bar chart.

# Create an array of positions for the first set of bars using the length of the x-axis categories.
p = np.arange(len(x))  # p = [0, 1, 2, 3], the positions for the first set of bars.

# Calculate positions for the second set of bars by adding the width of the bars to each position.
p1 = [j + width for j in p]  # p1 = [0.2, 1.2, 2.2, 3.2], shifted positions for the second set.

# Set labels for the axes and the title of the plot.
plt.xlabel("Programming Language", fontsize=10)  # Label for the x-axis with specified font size.
plt.ylabel("Marks", fontsize=10)  # Label for the y-axis with specified font size.
plt.title("Student Performance", fontsize=10)  # Title of the graph with specified font size.

# Plot the first set of bars:
plt.bar(p, y, width, color="r", label="Popularity")  
# - p: x-axis positions for the first set of bars.
# - y: heights of the first set of bars.
# - width: width of each bar.
# - color: color of the bars ('r' for red).
# - label: label used in the legend for this set of bars.

# Plot the second set of bars:
plt.bar(p1, z, width, color="y", label="Popularity1")  
# - p1: x-axis positions for the second set of bars.
# - z: heights of the second set of bars.
# - width: width of each bar.
# - color: color of the bars ('y' for yellow).
# - label: label used in the legend for this set of bars.

# Adjust the positions of the x-ticks to be in the center of the grouped bars:
plt.xticks(p + width / 2, x, rotation=20)  
# - p + width / 2: positions in the center of each grouped bar set.
# - x: labels for each position corresponding to programming languages.
# - rotation=20: rotate the labels by 20 degrees for better readability.

plt.legend()  # Display the legend on the plot to distinguish the datasets.
plt.show()  # Render and display the plot.
```

```python
import matplotlib.pyplot as plt  # Import the pyplot module from matplotlib for creating visualizations.

# Define the categories (programming languages) for the y-axis.
x = ["python", "C", "C++", "Java"]

# Define the data for the x-axis:
y = [85, 70, 60, 82]  # First dataset, e.g., marks or scores for each programming language.
z = [20, 30, 40, 50]  # Second dataset, e.g., another set of marks or different data.

# Set labels for the axes and the title of the plot.
plt.xlabel("Marks", fontsize=15)  # Label for the x-axis with specified font size.
plt.ylabel("Programming Language", fontsize=15)  # Label for the y-axis with specified font size.
plt.title("Student Performance", fontsize=15)  # Title of the graph with specified font size.

# Plot the first set of horizontal bars:
plt.barh(x, y, color="r", label="popularity")  
# - x: categories for the y-axis (programming languages).
# - y: lengths of the bars along the x-axis (first dataset).
# - color: color of the bars ('r' for red).
# - label: label used in the legend for this set of bars.

# Plot the second set of horizontal bars:
plt.barh(x, z, color="y", label="popularity1")  
# - x: categories for the y-axis (same as for the first set).
# - z: lengths of the bars along the x-axis (second dataset).
# - color: color of the bars ('y' for yellow).
# - label: label used in the legend for this set of bars.

plt.legend()  # Display the legend on the plot to distinguish between the datasets.
plt.show()  # Render and display the plot.
```
